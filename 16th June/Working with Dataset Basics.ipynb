Basic Exercises

```python
print("All Records:")
df.show()
print("Schema:\n")
df.printSchema()
print("First 10 records:")
df.show(10)
print("Product and price:")
df.select("product", "price").show()
print("City and Category:")
df.select("city", "category").show()
print("Total Records:", df.count())
print("Total Columns:", len(df.columns))
print("Distinct Cities:")
df.select("city").distinct().show()
print("Distinct Categories:")
df.select("category").distinct().show()
print("Distinct Status:")
df.select("status").distinct().show()
```

Filter Exercises

```python
print("Electronics Products:")
df.filter(df.category == "Electronics").show()
print("Furniture Products:")
df.filter(df.category == "Furniture").show()
print("Fashion Products:")
df.filter(df.category == "Fashion").show()
print("Orders from Hyderabad:")
df.filter(df.city == "Hyderabad").show()
print("Orders from Bangalore:")
df.filter(df.city == "Bangalore").show()
print("Completed Orders:")
df.filter(df.status == "Completed").show()
print("Pending Orders:")
df.filter(df.status == "Pending").show()
print("Cancelled Orders:")
df.filter(df.status == "Cancelled").show()
print("Price greater than 30000:")
df.filter(df.price > 30000).show()
print("Quantity greater than 2:")
df.filter(df.quantity > 2).show()

print("Electronics orders from Hyderabad:")
df.filter((df.category == "Electronics") & (df.city == "Hyderabad")).show()
print("Furniture orders from Delhi:")
df.filter((df.category == "Furniture") & (df.city == "Delhi")).show()
print("Fashion orders from Bangalore:")
df.filter((df.category == "Fashion") & (df.city == "Bangalore")).show()
print("Orders from Hyderabad or Mumbai:")
df.filter((df.city == "Hyderabad") | (df.city == "Mumbai")).show()
print("Orders where price is between 10000 and 50000:")
df.filter((df.price >= 10000) & (df.price <= 50000)).show()
```

Column Exercises


```python
df_total_amount = df.withColumn("total_amount", df.price * df.quantity)
print("Total Amount:")
df_total_amount.show()

df_tax = df_total_amount.withColumn("tax", df_total_amount.total_amount * 0.05)
print("Tax:")
df_tax.show()

df_final = df_tax.withColumn("total_amount_with_tax", df_tax.total_amount + df_tax.tax)
print("Total Amount with Tax:")
df_final.show()

df_discount = df_final.withColumn("discount", df_final.total_amount_with_tax * 0.1)
print("Discount:")
df_discount.show()

df_net_amount = df_discount.withColumn("net_amount", df_discount.total_amount_with_tax - df_discount.discount)
print("Net Amount:")
df_net_amount.show()

df = df.withColumnRenamed("price", "unit_price")
df = df.withColumnRenamed("status", "order_status")
print("After renaming price and status column:")
df.show()

print("After dropping category column:")
df.drop("category").show()
print("After dropping city column:")
df.drop("city").show()

df_copy = df_net_amount
print("Copy DataFrame with all calculated columns:")
df_copy.show()
```

GroupBy Exercises

```python
from pyspark.sql.functions import col, sum, avg, max, min
print("Orders by city:")
df.groupBy("city").count().show()
print("Orders by category:")
df.groupBy("category").count().show()
print("Orders by status:")
df.groupBy("order_status").count().show()
print("Total revenue by city:")
df.groupBy("city").agg(sum(col("quantity") * col("unit_price")).alias("total_revenue")).show()
print("Total revenue by category:")
df.groupBy("category").agg(sum(col("quantity") * col("unit_price")).alias("total_revenue")).show()
print("Total revenue by product:")
df.groupBy("product").agg(sum(col("quantity") * col("unit_price")).alias("total_revenue")).show()
print("Average product price by category:")
df.groupBy("category").agg(avg(col("unit_price")).alias("avg_price")).show()
print("Maximum product price by category:")
df.groupBy("category").agg(max(col("unit_price")).alias("max_price")).show()
print("Minimum product price by category:")
df.groupBy("category").agg(min(col("unit_price")).alias("min_price")).show()
print("Total quantity sold by product:")
df.groupBy("product").agg(sum(col("quantity")).alias("total_quantity")).show()
```

```python
from pyspark.sql.functions import *

df_price_band = df.withColumn(
    "price_band",
    when(col("unit_price")>=50000,"Premium")
    .when(col("unit_price")>=10000,"Standard")
    .otherwise("Budget")
)
df_price_band.show()
df_price_band.groupBy("price_band").count().show()

df_order_size = df.withColumn("order_size", 
    when(col("quantity")>=4, "Large").otherwise("Small")
)
df_order_size.show()
df_order_size.groupBy("order_size").count().show()

df_revenue_band = df_total_amount.withColumn("revenue_band",
    when(col("total_amount")>=100000,"High")
    .otherwise("Low"))
df_revenue_band.show()
```

Window Functions

```python
from pyspark.sql.functions import *
from pyspark.sql.window import Window

df_revenue = df.withColumn("revenue", col("quantity") * col("unit_price"))
w = Window.orderBy(col("revenue").desc())
df_revenue.withColumn("row_number", row_number().over(w)).show()
df_revenue.withColumn("rank", rank().over(w)).show()
df_revenue.withColumn("dense_rank", dense_rank().over(w)).show()
df_revenue.withColumn("rank",rank().over(w)).filter(col("rank")<=5).show()
df_revenue.orderBy(col("revenue").desc()).show(1)

w_city = Window.partitionBy("city").orderBy(col("revenue").desc())
df_revenue.withColumn("row_number", row_number().over(w_city)).filter(col("row_number")<=1).show()

w_category = Window.partitionBy("category").orderBy(col("revenue").desc())
df_revenue.withColumn("row_number", row_number().over(w_category)).filter(col("row_number")<=1).show()
df_revenue.withColumn("running_total", sum("revenue").over(Window.orderBy("order_id").rowsBetween(Window.unboundedPreceding,0))).show()

df_revenue.withColumn("prev_revenue", lag("revenue").over(w)).show()
df_revenue.withColumn("next_revenue", lead("revenue").over(w)).show()
df_revenue.withColumn("previous_revenue", lag("revenue", 1).over(w)) \
  .select("order_id", "revenue", "previous_revenue") \
  .show()

df_revenue.withColumn("previous_revenue", lag("revenue", 1).over(w)) \
  .withColumn(
      "revenue_difference", col("revenue") - col("previous_revenue")) \
  .select("order_id", "revenue", "previous_revenue","revenue_difference").show()
df_revenue.withColumn("rank", row_number().over(w_city)).filter(col("rank") <= 2).show()
df_revenue.withColumn("rank", row_number().over(w_category)).filter(col("rank") <= 3).show()
ranking_window = Window.orderBy(col("revenue").desc())

df_revenue.withColumn("revenue_rank", dense_rank().over(ranking_window)) \
  .select(
      "order_id",
      "product",
      "city",
      "revenue",
      "revenue_rank"
  ).orderBy("revenue_rank").show()
```

```python
sales_df = spark.read.option("header", True).option("inferSchema", True).csv("/FileStore/sales_data")
sales_df.show()

sales_df = sales_df.withColumn("revenue", col("quantity") * col("price"))
sales_df.show()
completed_df = sales_df.filter(col("status") == "Completed")
completed_df.show()
completed_df.agg(sum("revenue").alias("city_revenue")).show()
city_revenue = completed_df.groupBy("city").agg(sum("revenue").alias("city_revenue"))
city_revenue.show()

category_revenue = completed_df.groupBy("category").agg(sum("revenue").alias("category_revenue"))
category_revenue.show()

completed_df.write.mode("overwrite").parquet("/FileStore/output/sales_parquet")
parquet_df = spark.read.parquet("/FileStore/output/sales_parquet")
parquet_df.show()
parquet_df.createOrReplaceTempView("sales_parquet_view")
spark.sql("""SELECT city, category, COUNT(*) AS total_orders, SUM(revenue) AS total_revenue
FROM sales_parquet_view GROUP BY city, category ORDER BY total_revenue DESC""").show()
```
