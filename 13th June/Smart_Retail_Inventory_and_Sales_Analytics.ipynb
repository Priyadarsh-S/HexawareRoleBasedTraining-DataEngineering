```python
from pyspark.sql.functions import col, when
products_df.filter(col("supplier_id").isNull()).show()
```

```python
sales_df = sales_df.na.fill({"payment_mode": "Not Provided"})
sales_df.show()
```

```python
products_df = products_df.na.fill({"supplier_id": "UNKNOWN"})
products_df.show()
```

```python
products_df = products_df.withColumn("data_quality_status", when(
    col("supplier_id") == "UNKNOWN", "Missing Supplier ID").otherwise("Clean"))
products_df.show()

inventory_df = inventory_df.withColumn("data_quality_status", when(
    col("stock_quantity") == 0, "Missing Stock Quantity").otherwise("Clean"))
inventory_df.show()

sales_df = sales_df.withColumn("data_quality_status", when(
    (col("sale_amount") == 0) | (col("payment_mode") == "Not Provided"), "Incomplete"
    ).otherwise("Clean"))
sales_df.show()
```

Part 7: Window Functions


```python
from pyspark.sql.window import Window
product_revenue = sales_df.join(products_df, "product_id", "left") \
  .groupBy("product_id", "product_name", "category").agg(
    F.sum("sale_amount").alias("total_revenue"))
window = Window.orderBy(col("total_revenue").desc())
product_revenue.withColumn("rank", F.rank().over(window)).show()
```

```python
stores_revenue = sales_df.groupBy("store_id").agg(
  F.sum("sale_amount").alias("total_revenue"))
window = Window.orderBy(col("total_revenue").desc())
stores_revenue.withColumn("rank", F.rank().over(window)).show()
```

```python
window = Window.partitionBy("category").orderBy(F.desc("total_revenue"))
product_revenue = product_revenue.withColumn("category_rank", F.rank().over(window)) \
    .orderBy("category","category_rank")
product_revenue.show()
```

```python
window = Window.partitionBy("category").orderBy(F.desc("total_revenue"))
product_revenue.withColumn("category_rank", F.rank().over(window)) \
  .filter(col("category_rank") == 1).show()
```

```python
window = Window.partitionBy("category").orderBy(F.desc("total_revenue"))
product_revenue.withColumn("category_rank", F.dense_rank().over(window)) \
  .filter(col("category_rank") <= 3).orderBy("category", "category_rank").show()
```

```python
store_state_revenue = sales_df.join(stores_df, "store_id","left")\
    .groupBy("store_id","store_name","state")\
    .agg(F.sum("sale_amount").alias("total_revenue"))
w_state = Window.partitionBy("state").orderBy(F.desc("total_revenue"))
store_state_revenue.withColumn("state_rank", F.rank().over(w_state))\
    .filter(F.col("state_rank") == 1).show()
```

```python
running_revenue = sales_df.groupBy("sale_date").agg(
    F.sum("sale_amount").alias("total_revenue")).orderBy("sale_date")
window = Window.orderBy("sale_date")
running_revenue.withColumn("running_total", F.sum("total_revenue").over(window)).show()
```

```python
running_revenue = sales_df.groupBy("sale_date").agg(
    F.sum("sale_amount").alias("daily_revenue")).orderBy("sale_date")
window = Window.orderBy("sale_date")
running_revenue.withColumn("prev_revenue", F.lag("daily_revenue").over(window)) \
    .withColumn("diff_prev_revenue", col("daily_revenue")-col("prev_revenue")).show()
```

```python
window = Window.orderBy("sale_date")
sales_df.withColumn("next_amount", F.lead("sale_amount", 1).over(window)).show()
```

```python
window = Window.partitionBy("product_id").orderBy("sale_date")
sales_df.withColumn("prev_amount", F.lag("sale_amount", 1).over(window)) \
    .filter(col("sale_amount") > col("prev_amount")).show()
```

```python
spark.sql("""
    SELECT i.inventory_id, i.store_id, i.product_id, p.product_name,
           i.stock_quantity, i.reorder_level,
           CASE WHEN i.stock_quantity <= i.reorder_level
                THEN 'Reorder Required' ELSE 'Sufficient Stock' END AS stock_status
    FROM inventory i
    JOIN products p ON i.product_id = p.product_id
    WHERE i.stock_quantity <= i.reorder_level
""").show()
```

Part 10: Final Gold Reports


```python
final_sales = spark.read.parquet("silver/sales")
store_perf = final_sales.join(stores_df.select("store_id","store_name","city","state"), "store_id","left")\
    .groupBy("store_id","store_name","city","state")\
    .agg(F.count("sale_id").alias("total_sales"), F.sum("sale_amount").alias("total_revenue")
    ).orderBy("total_revenue", ascending=False)
store_perf.show()
store_perf.write.mode("overwrite").parquet("gold/report_store_performance")
```

```python
product_perf = final_sales.join(products_df.select("product_id","product_name","category","brand"), "product_id","left")\
    .groupBy("product_id","product_name","category","brand")\
    .agg(F.sum("quantity_sold").alias("total_quantity_sold"), F.sum("sale_amount").alias("total_revenue")
    ).orderBy("total_revenue", ascending=False)
product_perf.show()
product_perf.write.mode("overwrite").parquet("gold/report_product_performance")
```
