Part 2: Data Cleaning


```python
from pyspark.sql.functions import col, when
customers_df.filter(col("plan_id").isNull()).show()
```

```python
payments_df.na.fill({"payment_mode": "Not Provided"}).show()
```

```python
customers_clean_df = customers_df.withColumn("data_quality_status", when(
    col("plan_id").isNull(), "Incomplete").otherwise("Complete"))
customers_clean_df.show()

usage_clean_df = usage_df.withColumn("data_quality_status", when(
    col("data_used_gb").isNull(), "Incomplete").otherwise("Complete"))
usage_clean_df.show()

payments_clean_df = payments_df.withColumn("data_quality_status", when(
    col("amount_paid").isNull()|(col("payment_mode").isNull()), "Incomplete"
    ).otherwise("Complete"))
payments_clean_df.show()
```

Part 5: Transformations


```python
usage_category_df = usage_df.withColumn("usage_category", when(
    col("data_used_gb") >= 70, "Heavy User").when(
    (col("data_used_gb") >= 30), "Medium User"
    ).otherwise("Low User"))
usage_category_df.show()
```

```python
payment_category_df = payments_df.withColumn("payment_category", when(
    col("amount_paid") >= 1000, "High Payment").when(
    col("amount_paid") >= 500, "Medium Payment"
    ).otherwise("Low Payment"))
payment_category_df.show()
```

```python
from pyspark.sql.functions import when, col, trim

complete_data_df = complete_data_df.withColumn(
    "churn_risk",
    when(
        (trim(col("status")) == "Inactive") |
        (trim(col("payment_status")) != "Success"),
        "High Risk"
    )
    .when(col("data_used_gb") < 15, "Medium Risk")
    .otherwise("Low Risk")
)
complete_data_df.show()
```

```python
complete_data_df = complete_data_df.withColumn("over_usage_gb",
      col("data_used_gb") - col("data_limit_gb"))
complete_data_df.show()
```

```python
complete_data_df = complete_data_df.withColumn("over_usage_flag", when(
    col("over_usage_gb") > 0, "Yes").otherwise("No"))
complete_data_df.show()
```

```python
customer_usage_summary_report = complete_data_df.select(
    "customer_id",
    "customer_name",
    "city",
    "plan_name",
    "usage_month",
    "data_used_gb",
    "data_limit_gb",
    "over_usage_flag",
    "amount_paid",
    "payment_status",
    "churn_risk"
)

customer_usage_summary_report.show(truncate=False)
```

```python
from pyspark.sql.functions import countDistinct
plan_performance_report = complete_data_df.groupBy(
    "plan_name"
).agg(
    countDistinct("customer_id").alias("total_customers"),
    sum("data_used_gb").alias("total_data_usage"),
    avg("data_used_gb").alias("average_data_usage"),
    sum("amount_paid").alias("total_revenue")
)
plan_performance_report.show()
```

```python
city_revenue_report = complete_data_df.groupBy("city").agg(
    countDistinct("customer_id").alias("total_customers"),
    sum("amount_paid").alias("total_revenue"),
    avg("amount_paid").alias("average_revenue"))
city_revenue_report.show()
```
