Part 7 - Window Functions


```python
from pyspark.sql.window import Window
from pyspark.sql.functions import *

patients_fee_df = appointments_df.groupBy("patient_id").agg(sum("consultation_fee")
    .alias("total_consultation_fee"))
window_spec = Window.orderBy(col("total_consultation_fee").desc())
patients_fee_df.withColumn("rank", rank().over(window_spec)).show()
```

```python
window_spec = Window.orderBy(col("total_consultation_fee").desc())
patients_fee_df.withColumn("dense_rank", dense_rank().over(window_spec)).show()
```

```python
window_spec = Window.orderBy(col("total_consultation_fee").desc())
patients_fee_df.withColumn("row_number", row_number().over(window_spec)).show()
```

```python
patients_fee_df.orderBy(desc("total_consultation_fee")).show(1)
```

```python
patients_fee_df.orderBy(desc("total_consultation_fee")).show(3)
```

```python
patients_spending_city = patients_df.join(patients_fee_df, "patient_id", "inner")

citywindow = Window.partitionBy("city").orderBy(col("total_consultation_fee").desc())
patients_spending_city.withColumn("rank",
      rank().over(citywindow)).filter(col("rank")==1).show()
```

```python
citywindow = Window.partitionBy("city").orderBy("total_consultation_fee")
patients_spending_city.withColumn("rank",
      rank().over(citywindow)).filter(col("rank")==1).show()
```

```python
window_spec = Window.orderBy("patient_id")
patients_fee_df.withColumn("running_total",
          sum("total_consultation_fee").over(window_spec)).show()
```

```python
patients_fee_df.withColumn("next_fee", lead("total_consultation_fee")
          .over(window_spec)).show()
```

```python
patients_fee_df.withColumn("previous_fee", lag("total_consultation_fee")
          .over(window_spec)).show()
```

Part 10 - ETL Project


```python
patients_df = spark.read.csv("patients.csv", header=True, inferSchema=True)
appointments_df = spark.read.csv("appointments.csv", header=True, inferSchema=True)
```

```python
preferences_df = spark.read.option("multiLine", "true") \
  .json("patient_preferences.json")
```

```python
fill_patients_df = patients_df.na.fill({"city": "Unknown",
    "blood_group": "Not Available"}).show()

fill_appointments_df = appointments_df.na.fill({"consultation_fee": 0}).show()
```

```python
all_df = patients_df.join(appointments_df, "patient_id", "full") \
  .join(preferences_df, "patient_id", "full")
all_df.show()
```

```python
age_group_df = patients_df.withColumn("age_group", when(
    col("age") < 30, "Young").when(col("age") < 50, "Middle Age")
    .otherwise("Senior")
)
age_group_df.show()
```

```python
dept_revenue = appointments_df.groupBy("department") \
  .agg(sum("consultation_fee").alias("Revenue"))
dept_revenue.show()
```

```python
patients_spending_df = all_df.groupBy("patient_id", "patient_name") \
  .agg(sum("consultation_fee").alias("Total_Consultation_Fee"))
patients_spending_df.show()
```

```python
appointments_df.groupBy("department").agg(sum("consultation_fee") \
          .alias("Department Revenue")).show()
```

```python
all_df.write.mode("overwrite").parquet("final_data.parquet")
```
