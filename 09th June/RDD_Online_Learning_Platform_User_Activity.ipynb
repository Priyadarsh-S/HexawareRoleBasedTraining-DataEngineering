Real RDD Analytics


```python
activities_cnt = rdd.countByValue()
print(max(activities_cnt.items(), key=lambda x: x[1]))
```

```python
print(min(activities_cnt.items(), key=lambda x: x[1]))
```

```python
login = (activities_cnt["LOGIN"]/rdd.count())*100
print(login)
```

```python
video_play = (activities_cnt["VIDEO_PLAY"]/rdd.count())*100
print(video_play)
```

```python
print("Activity Report")
print("Total Activities:", rdd.count())
print("Unique Activities:", rdd.distinct().count())
```

```python
print("Unique Activity Report")
print("Unique Activities:",rdd.distinct().count())
print(rdd.distinct().collect())
```

```python
print("Frequency Report")
for k,v in activities_cnt.items():
  print(k,v)
```

```python
rdd.distinct().saveAsTextFile("distinct_activities")
```

```python
freq_rdd = spark.sparkContext.parallelize(
    [(k,v) for k,v in activities_cnt.items()]
)
freq_rdd.saveAsTextFile("activity frequencies")
```

```python
print("--- Learning Platform Summary ---")
print("Total Records:", rdd.count())
print("Unique Activities:", rdd.distinct().count())
print("Most Common:", max(activities_cnt.items(), key=lambda x: x[1]))
print("Least Common:", min(activities_cnt.items(), key=lambda x: x[1]))
print("Login Percentage:", round(login,2))
print("Video Play Percentage:", round(video_play,2))
```
