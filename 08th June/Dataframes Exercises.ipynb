```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Dataframes Exercises") \
    .getOrCreate()
```

```python
data = [
  (1, "Rahul Sharma", "Hyderabad", "Electronics", "Laptop", 1, 65000, "Delivered"),
  (2, "Priya Reddy", "Bangalore", "Electronics", "Mobile", 2, 25000, "Delivered"),
  (3, "Amit Kumar", "Mumbai", "Furniture", "Chair", 4, 3500, "Pending"),
  (4, "Sneha Patel", "Chennai", "Electronics", "Tablet", 1, 30000, "Cancelled"),
  (5, "Farhan Ali", "Delhi", "Furniture", "Table", 2, 12000, "Delivered"),
  (6, "Neha Singh", "Hyderabad", "Fashion", "Shoes", 3, 2500, "Delivered"),
  (7, "Arjun Verma", "Pune", "Fashion", "Watch", 1, 8000, "Pending"),
  (8, "Meera Nair", "Bangalore", "Electronics", "TV", 1, 45000, "Delivered"),
  (9, "Kiran Rao", "Hyderabad", "Furniture", "Desk", 1, 15000, "Delivered"),
  (10, "Ravi Kumar", "Mumbai", "Electronics", "Mobile", 1, 28000, "Delivered")
]
```

```python
columns = [
    "order_id",
    "customer_name",
    "city",
    "category",
    "product",
    "quantity",
    "price",
    "order_status"
]
```

```python
df = spark.createDataFrame(data, columns)
df.show()
```

```python
df_total_amount = df.withColumn(
    "total_amount",
    col("quantity") * col("price")
)
df_total_amount.show()
```

```python
df_discount = df_total_amount.withColumn(
    "discount",
    col("total_amount") * 0.1
)
df_discount.show()
```

```python
df_final_amount = df_discount.withColumn(
    "final_amount",
    col("total_amount") - col("discount")
)
df_final_amount.show()
```

```python
from pyspark.sql.functions import when
df_order_value_category = df_total_amount.withColumn(
    "order_value_category",
    when(col("total_amount") >= 50000, "High")
    .when(col("total_amount") >= 20000, "Medium")
    .otherwise("Low")
)
df_order_value_category.show()
```

```python
df_order_value_category.groupBy("order_value_category").count().show()
```

```python
final_report = df_final_amount.select(
    col("order_id").alias("Order ID"),
    col("customer_name").alias("Customer Name"),
    col("city").alias("City"),
    col("category").alias("Category"),
    col("product").alias("Product"),
    col("quantity").alias("Quantity"),
    col("price").alias("Price"),
    col("total_amount").alias("Final Amount"),
    col("discount").alias("Discount"),
    col("final_amount").alias("Final Amount"),
    col("order_status").alias("Order Status")
)
final_report.show()
```
