```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("PySpark Day 1") \
    .getOrCreate()
```

```python
data = [
    (101, "Rahul Sharma", "Data Engineering", 75000, "Hyderabad"),
    (102, "Priya Reddy", "AI Engineering", 85000, "Bangalore"),
    (103, "Amit Kumar", "Data Engineering", 65000, "Mumbai"),
    (104, "Sneha Patel", "Data Science", 95000, "Chennai"),
    (105, "Farhan Ali", "Cloud Engineering", 80000, "Delhi")
]
```

```python
columns = [
    "employee_id",
    "name",
    "department",
    "salary",
    "city"
]
```

```python
df = spark.createDataFrame(data, columns)
df.show()
```

```python
df.printSchema()
```

```python
df.select("name", "salary").show()
```

```python
df.filter(df.salary > 80000).show()
```

```python
df.filter(df.city == "Hyderabad").show()
```

```python
df.orderBy("salary").show()
```

```python
df.orderBy(df.salary.desc()).show()
```

```python
from pyspark.sql.functions import col

df_bonus = df.withColumn(
    "bonus",
    col("salary") * 0.10
)
df_bonus.show()
```

```python
df_renamed = df.withColumnRenamed(
    "salary",
    "monthly_salary"
)
df_renamed.show()
```

```python
df.groupBy("department").count().show()
```

```python
from pyspark.sql.functions import when
df_salary_category = df.withColumn("salary_category",
                                   when(col("salary") >= 85000, "High")
                                   .when(col("salary") >= 75000, "Medium")
                                   .otherwise("Low"))
df_salary_category.show()
```

```python
df_salary_category.groupBy("salary_category").count().show()
```

```python
final_report = df.withColumn("Bonus", col("salary") * 0.10) \
    .withColumn("Annual Salary", col("salary") * 12) \
    .select(
        col("employee_id").alias("Employee ID"),
        col("name").alias("Name"),
        col("department").alias("Department"),
        col("salary").alias("Salary"),
        col("Bonus"),
        col("Annual Salary")
    )
final_report.show()
```
