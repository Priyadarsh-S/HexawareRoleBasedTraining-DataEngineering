```python
df = spark.read.csv("airline_bookings.csv", header=True, inferSchema=True)
df.show()
df.printSchema()
```

```python
from pyspark.sql import functions as F
df.filter(col("status") == "Confirmed") \
  .agg(F.sum("ticket_price")).alias("Total Revenue").show()
```

```python
df.groupBy("airline") \
  .agg(F.avg("ticket_price").alias("Average Price")).show()
```

```python
df.agg(F.max("ticket_price").alias("Highest ticket price")).show()
```

```python
df.agg(F.min("ticket_price").alias("Lowest ticket price")).show()
```

```python
df.orderBy(col("ticket_price").desc()).show()
```

```python
df_tax = df.withColumn("tax", col("ticket_price")*0.05).show()
```

```python
df_final_price = df.withColumn(
    "tax", col("ticket_price")*0.05
).withColumn(
    "final_price", col("ticket_price") + col("tax")
).show()
```

```python
from pyspark.sql.functions import when
df_price_category = df.withColumn(
    "price_category", when(col("ticket_price") >= 30000, "Premium")
    .when(col("ticket_price") >= 10000, "Standard")
    .otherwise("Budget")
).show()
```

```python
df_price_category = df.withColumn(
    "price_category", when(col("ticket_price") >= 30000, "Premium")
    .when(col("ticket_price") >= 10000, "Standard")
    .otherwise("Budget")
).groupBy("price_category").count().show()
```

```python
df.filter(col("status") == "Confirmed") \
  .write.mode("overwrite").parquet("confirmed_flights.parquet")
```
