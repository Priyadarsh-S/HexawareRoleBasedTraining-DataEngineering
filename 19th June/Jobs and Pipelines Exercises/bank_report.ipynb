```python
account_data = [
    ("ACC101", "Checking", 5000.00, "New York"),
    ("ACC102", "Savings", 12000.50, "Chicago"),
    ("ACC103", "Checking", 3200.00, "New York"),
    ("ACC104", "Savings", 15000.00, "San Francisco"),
    ("ACC105", "Checking", 850.00, "Chicago")
]

schema = ["account_id", "account_type", "balance", "city"]
account_df = spark.createDataFrame(account_data, schema=schema)

from pyspark.sql.functions import *
report_df = account_df.groupBy("city").agg(sum("balance").alias("total_balance"))
display(report_df)
```
