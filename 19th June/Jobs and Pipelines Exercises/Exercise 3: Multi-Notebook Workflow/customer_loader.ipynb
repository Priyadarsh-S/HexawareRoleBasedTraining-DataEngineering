```python
customer_data = [
    (1, "John Doe", "New York"),
    (2, "Jane Doe", "Chicago"),
    (3, "Sam Spade", "New York"),
    (4, "Miles Archer", "San Francisco"),
    (5, "Mary Jane", "Chicago")
]

schema = ["customer_id", "customer_name", "city"]
customer_df = spark.createDataFrame(customer_data, schema=schema)

customer_df.write.mode("overwrite").saveAsTable("customer_table")
print("Customer data successfully loaded and saved to customer_table.")
```
