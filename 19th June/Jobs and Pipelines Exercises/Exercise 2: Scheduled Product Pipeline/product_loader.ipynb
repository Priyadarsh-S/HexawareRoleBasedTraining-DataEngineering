```python
product_data = [
    ("Laptop", "Electronics", 1200.00),
    ("Smartphone", "Electronics", 800.00),
    ("Desk Chair", "Furniture", 150.00),
    ("Monitor", "Electronics", 300.00),
    ("Coffee Maker", "Appliances", 85.00)
]
schema = ["product_name", "category", "price"]
product_df = spark.createDataFrame(product_data, schema=schema)
display(product_df)
```

```python
product_df.createOrReplaceTempView("products")
avg_price = spark.sql("SELECT category, avg(price) AS average_price FROM products GROUP BY category")
display(avg_price)
```
